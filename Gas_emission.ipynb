```python
import os
import pandas as pd
os.chdir(r'/Users/anceaumathilde/data-ft-par-labs/Statistically_Significant_Insights/gas_source')
df = pd.read_csv('Final_Gas_Emission.csv')
```

```python
select_col=['Germany','Italy','Austria','Belgium','Lithuania','Bulgaria','Luxembourg','Cyprus','Malta','Coratia','Netherlands','Denmark','Poland','Spain','Portugal','Estonia','Romania','Finland','Slovakia','France','Slovenia','Greece','Ireland','Sweden','Hungary']
```

```python
frames=[]

for i in select_col:
    data=df.loc[df.Country==i]
    frames.append(data)
ue = pd.concat(frames)           
ue['State']='EU'
```

```python
sel_col2=['South Africa','Turkey','Saudi Arabia','Argentina','Australia','Brazil','Canada','South Korea','United States','India','Indonesia','Japan','Mexico','United Kingdom','Russia','China','Turkey']
```

```python
frames=[]

for i in sel_col2:
    data=df.loc[df.Country==i]
    frames.append(data)
country = pd.concat(frames)

ue
```

```python

ue.columns
ue=ue[['State','Country', 'Gas', '2018', '2017',
       '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008',
       '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', '1999',
       '1998', '1997', '1996', '1995'
       ]]
```

```python
df3=ue.pivot_table(index=['Country','State'], columns='Gas')
ue_2=df3.stack(0)
ue_2=pd.DataFrame(ue_2.to_records())
ue_2.rename(columns={'level_0':'Date'},inplace = True)
del ue_2['index']
```

```python
country=country[['Country', 'Gas', '2018', '2017',
       '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008',
       '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', '1999',
       '1998', '1997', '1996', '1995'
       ]]

df4=country.pivot_table(index=['Country'], columns='Gas')
country=df4.stack(0)
country=pd.DataFrame(country.to_records())
country.rename(columns={'level_0':'Date'},inplace = True)
del country['index']
```

```python
country['State']='Not EU'
```

```python
eu_3=ue_2.groupby('Date').sum()
eu_3=pd.DataFrame(eu_3.to_records())
eu_3['Country']='EU'
```

```python
gas_glob=pd.concat([country,ue_2],axis=0)
gas_EUW=pd.concat([country,eu_3],axis=0)

gas_EUW.to_csv('Gas_EUvsW.csv')
gas_glob.to_csv('Gas_global.csv')
```
